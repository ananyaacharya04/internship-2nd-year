```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([0,2,3,6,7,10])
ypoints=np.array([0,5,11,12,20,25])
plt.figure(figsize=(3,2))
plt.plot(xpoints,ypoints,"r")

plt.title("graph")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.bar(xpoints,ypoints,color="purple")
plt.title("graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(xpoints)
plt.show()
```

```python
plt.scatter(xpoints,ypoints,color="black")
plt.title("graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
firms=[1,2,3,4,5,6]
market=[20,25,30,35,40,45]
explode=[0,0,0,0.5,0,0]
plt.pie(market,labels=firms,startangle=45,shadow=True)
plt.legend(title="firms")
plt.show()
```

```python
plt.hist(xpoints,ypoints)
plt.title("graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.subplot(3,2,1)
plt.plot(x,y)

x=np.array([1,2,3,4,5])
y=np.array([5,15,25,35,45])
plt.subplot(3,2,2)
plt.plot(x,y,"ko")

x=np.array([1,2,3,4,5])
y=np.array([5,15,25,35,45])
plt.subplot(3,2,3)
plt.plot(x,y,"r*")

x=np.array([1,2,3,4,5])
y=np.array([5,15,25,35,45])
plt.subplot(3,2,4)
plt.plot(x,y,"purple")

x=np.array([1,2,3,4,5])
y=np.array([5,15,25,35,45])
plt.subplot(3,2,5)
plt.plot(x,y,"y^")

x=np.array([1,2,3,4,5])
y=np.array([5,15,25,35,45])
plt.subplot(3,2,6)
plt.plot(x,y,"orange")

plt.show()
```
